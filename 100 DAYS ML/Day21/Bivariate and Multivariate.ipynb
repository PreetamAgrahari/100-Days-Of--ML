SCATTERPLOT (NUMERICAL-NUMERICAL)

```python
sns.scatterplot(data=tips, x="total_bill", y="tip",hue="sex",style="smoker",size="size")
```

BAR PLOT (NUMERICAL-CATAGORICAL)

```python
titanic.head()
```

```python
sns.barplot(data=titanic,x="Pclass",y="Age",hue="Sex")
```

BOX PLOT (NUMERICAL-CATAGORICAL)

```python
sns.boxplot(data=titanic,x="Sex",y="Age",hue="Survived")
```

DISTPLOT (NUMERICAL-CATAGORICAL)

```python
#sns.histplot(data=titanic, x="Age", kde=True)
sns.histplot(titanic[titanic["Survived"] == 0]["Age"],kde=True, alpha=0)
sns.histplot(titanic[titanic["Survived"] == 1]["Age"],kde=True, alpha=0)
```

HEATMAP PLOT (CATAGORICAL-CATAGORICAL)

```python
titanic.head(3)
```

```python
sns.heatmap(pd.crosstab(titanic["Pclass"],titanic["Survived"]))
```

```python
#titanic.groupby("Pclass")["Survived"].mean()*100

# (titanic.groupby("Pclass")["Survived"].mean()*100).plot(kind="bar")

# (titanic.groupby("Sex")["Survived"].mean()*100)

(titanic.groupby("Embarked")["Survived"].mean()*100)
```

CLUSTERMAP (CATAGORICAL-CATAGORICAL)

```python
# sns.clustermap(pd.crosstab(titanic["SibSp"],titanic["Survived"]))

sns.clustermap(pd.crosstab(titanic["Parch"],titanic["Survived"]))
```
